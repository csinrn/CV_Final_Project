# Frequency method
Use STFT to determine fingerprint frequency.

In Frequency enhancement method, we go through the following algorithm:
1. Generate regional *frequency image* by stft
2. Generate regional *orientation image* by sobel
3. Generate *region mask*
4. Simulate fingerprint by Surface Wave Model: `i(x, y) =  A*cos[2*pi*f(xcos(theta)+ ysin(theta)}]`

First, we import the necessary library and read in the original fingerprint image

```python
import cv2
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread('./pic/1.bmp')
if pic.shape[2] == 3:
    pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
    
# TODO: histogram equalization

plt.imshow(pic, cmap='gray')
```

# Frequency Image
To generate the frequency image, we go through the following steps:
1. Set the window for stft, in this case, stftseg=16
2. Calculate 2d stft transform. We calculate row and col repectively
3. Calculate the mean stft in each 16*16 stft window as the final frequency
    
The result frequency image as following. We can see that in the middle and left parts of this fingerprint image are of higher frequency, which fit the original image.

```python
from scipy.signal import stft
import numpy as np
import math

res = []
fs = 10e3
stftseg = 16

print("original pic shape: ", pic.shape)

# calculate col stft into res
for i in range(pic.shape[0]):
    f, t, Zxx = stft(pic[i], fs, nperseg=stftseg)
    # print(f.shape, t.shape)
    res.append(Zxx)
res = np.array(res)
res = np.abs(res)
print("shape of col stft: ", res.shape)

f = []
res2 = []
# calculate row stft into res2
for i in range(pic.shape[1]):
    f, t, Zxx = stft(pic[:,i], fs, nperseg=stftseg)
    res2.append(Zxx)
res2 = np.array(res2)
res2 = np.abs(res2)
print("shape of row stft: ", res2.shape)

# calculate mean grid cols stft
step = math.ceil(pic.shape[0]/res2.shape[2])
res = [sum(res[i: i+step]) for i in range(0, len(res), step)]
res = np.array(res) /step/step
print("shape of gridded col stft: ", res.shape)

# calculate mean grid rows stft
step = math.ceil(pic.shape[1]/res.shape[2])
res2 = [sum(res2[i: i+step]) for i in range(0, len(res2), step)]
res2 = np.array(res2) /step/step
print("shape of gridded row stft: ", res2.shape)

# prune the margin of picture 
res = res[:-1,:, 1:-1]
res2 = res2[:-1,:, 1:-1]

print("shape of pruned col stft: ", res.shape)
print("shape of pruned col stft: ", res2.shape)

# combine row and col stft value
res = (res + res2.T)/2
# calculate pdf of res
res_ = []
for i in range(res.shape[0]):
    for j in range(res.shape[2]):
        ssum = sum(res[i,:,j])
        temp = 0
        for k in range(len(f)):
            # use mean of frequency as the final frequency
            temp = temp + f[k] * res[i,k,j]/ssum
        res_.append(temp)
freq = np.array(res_).reshape((res.shape[0], res.shape[2]))
print("shape of frequency image:", freq.shape)
print('list of frequency', f)
plt.title('Frequency Image of seg='+str(stftseg))
plt.imshow(freq)
```

# Orientation image
To generate the following image, we go through the following steps:
1. extract Gx and Gy by Sobel, which is the x direction and y direction gradient of the image
2. Calculate theta = Gx/Gy
3. Calculate the mean theta in grid 16*16 to pair with the stft grid

In the following images, we can see that in the top side of fingerprint, angle is close to zero(horizontal), and in the bottom side, the angle become larger (vertical) 

```python
# sobel kernel size
kernel_size = 17

# extract Gx and Gy by Sobel
plt.subplot(151)
gx = cv2.Sobel(pic, -1, 2, 0, ksize=kernel_size, borderType=cv2.BORDER_DEFAULT)
plt.title('Gx')
plt.imshow(gx, cmap = 'magma')

plt.subplot(152)
gy = cv2.Sobel(pic, -1, 0, 2, ksize=kernel_size, borderType=cv2.BORDER_DEFAULT)
plt.title('Gy')
plt.imshow(gy, cmap = 'magma')

'''
# another equation to extract theta, but did not show a good result in this case
stepr, stepc = math.ceil(pic.shape[0]/freq.shape[0]), math.ceil(pic.shape[1]/freq.shape[1])
print(stepr, stepc)
gxy = 2*gx*gy
gxx = gx*gx - gy*gy
print('gxy shape:', gxy.shape)
r,c = int(gxy.shape[0]/stepr), int(gxy.shape[1]/stepc)
'''

# calculate theta
theta = np.arctan(np.nan_to_num(gy/gx))
plt.subplot(153)
plt.title('original theta')
plt.imshow(theta, cmap='magma')

# calculate mean theta in 16*16 grid
res_ = []
for i in range(0, theta.shape[0], int(stftseg/2)):
    t = []
    for j in range(0, theta.shape[1], int(stftseg/2)):
        t.append(theta[i:i+stftseg, j:j+stftseg].sum())
    res_.append(np.array(t))
theta = np.array(res_).reshape((freq.shape[0]+1, freq.shape[1]+1))

theta = theta[0:-1, 0:-1]
theta = theta/stftseg/stftseg
print('theta shape', theta.shape)

plt.subplot(155)
plt.title('mean theta in grid 16*16')
plt.imshow(theta, cmap='magma')
```

# Region Mask
To avoid the unecessary contour, we create a mask for where fingerprint is located. And visualize the result image based on this mask.

```python
ret, region = cv2.threshold(pic, 150, 255, cv2.THRESH_BINARY_INV)
rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
region = cv2.dilate(region, rect, iterations = 1)

plt.title('Region mask')
plt.imshow(region)
```

# Simulation
i(x, y) =  Acos[ 2*pi*f{xcos(theta)+ ysin(theta)}]

```python
A = 1

res2 = []
for i in range(0, 20):
    for j in range(0, 20):
        f, th = 0.1, 0.785
        v = A* math.cos( 2*math.pi * f * (j*math.cos(th) + i*math.sin(th)))
        # i = 0 if i<0 else i
        res2.append(v)

res2 = np.array(res2).reshape((20, 20))
plt.subplot(121)
plt.title('f=1, angle=17.9')
plt.imshow(res2)

res2 = []
for i in range(0, 20):
    for j in range(0, 20):
        f, th = 500, 0.785
        v = A* math.cos( 2*math.pi * f * (j*math.cos(th) + i*math.sin(th)))
        # i = 0 if i<0 else i
        res2.append(v)

res2 = np.array(res2).reshape((20, 20))
plt.subplot(122)
plt.imshow(res2)
```

```python
A = 100

res = []

stepr, stepc = pic.shape[0]/freq.shape[0], pic.shape[1]/freq.shape[1]
for i in range(0, pic.shape[0]):
    for j in range(0, pic.shape[1]):
        gridr, gridc = math.floor(i/stepr), math.floor(j/stepc)

        if region[i][j] == 0:
            res.append(0)
            continue
        f, th = freq[gridr][gridc], theta[gridr][gridc]
        if (gridr+1 <freq.shape[0]) and (gridc+1 <freq.shape[1]):
            f, th = (f + freq[gridr+1][gridc+1])/2, (th +theta[gridr+1][gridc+1])/2
        f = f
        v = A* math.cos( 2*math.pi * f * (j*math.cos(th) + i*math.sin(th)))
        # i = 0 if i<0 else i
        res.append(v)

res = np.array(res).reshape(pic.shape)
print(res)
plt.title('Final image')
plt.imshow(res)
```
